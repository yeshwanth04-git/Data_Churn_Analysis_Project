```python
customer_df=spark.read.option("header","true").option("inferSchema","true").csv("/Volumes/bank_analytics_project/bronze/raw_data/customer_data.csv")
account_df=spark.read.option("header","true").option("inferSchema","true").csv("/Volumes/bank_analytics_project/bronze/raw_data/account_data.csv")
```

```python
df_joined = customer_df.join(account_df, "CustomerId", "inner") \
    .select(
        customer_df["CustomerId"],
        customer_df["Surname"],
        customer_df["CreditScore"],
        customer_df["Geography"],
        customer_df["Gender"],
        customer_df["Age"],
        customer_df["Tenure"],
        customer_df["EstimatedSalaryINR"],
        account_df["NumOfProducts"],
        account_df["HasCrCard"],
        account_df["IsActiveMember"],
        account_df["Exited"],
        account_df["BalanceINR"]
    )
```

```python
df_joined.write.format("delta").mode("overwrite").saveAsTable("bank_analytics_project.bronze.Bank")
```
