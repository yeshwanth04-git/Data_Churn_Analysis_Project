```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
df_silver=spark.table("bank_analytics_project.bronze.bank")
```

```python
df_silver=df_silver.withColumn("Ingestion_time",current_timestamp())
```

```python
fill_values = {
    "Surname": "Unknown",
    "Geography": "Unknown",
    "Gender": "Unknown",
    "HasCrCard": "No",
    "IsActiveMember": "No"
}

df_silver = df_silver.fillna(fill_values)
```

```python
df_silver = df_silver.dropDuplicates()
```

```python
for col_name in df_silver.columns:
    df_silver = df_silver.withColumn(col_name, trim(df_silver[col_name]) if dict(df_silver.dtypes)[col_name] == 'string' else df_silver[col_name])
```

```python
df_silver.write.format("delta").mode("overwrite").option("overwriteSchema","true").saveAsTable("bank_analytics_project.silver.Bank")
```
