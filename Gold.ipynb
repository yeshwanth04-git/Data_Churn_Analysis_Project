Have more than 100000 balance and exited

```python
%sql
CREATE TABLE if not exists bank_analytics_project.gold.exited_high_balance
USING DELTA
AS
SELECT *
FROM bank_analytics_project.silver.bank
WHERE BalanceINR > 100000 AND Exited = 1;
```

Have credit score more than 600 and have more than 1 product,
and are still with the bank.

```python
%sql
CREATE TABLE IF NOT EXISTS bank_analytics_project.gold.active_high_credit_multi_product
USING DELTA
AS
SELECT *
FROM bank_analytics_project.silver.bank
WHERE CreditScore > 600 AND NumOfProducts > 1 AND Exited = 0;
```

Customers who are active
and having balance more than 1,20,000 and are with
the bank for more than 3 years.

```python
%sql
CREATE TABLE IF NOT EXISTS bank_analytics_project.gold.active_female_high_balance_long_term
USING DELTA
AS
SELECT *
FROM bank_analytics_project.silver.bank
WHERE Exited = 0
  AND BalanceINR > 120000
  AND Gender = 'Female'
  AND Tenure > 3;
```

Based on geography need to find customers who are with the
bank for more than 5 years, dividing on gender.

```python
%sql
CREATE TABLE IF NOT EXISTS bank_analytics_project.gold.long_term_by_geography_gender
USING DELTA
AS
SELECT Geography, Gender, count(distinct CustomerId) as Customer_count
FROM bank_analytics_project.silver.bank
WHERE Tenure > 5
group by Geography, Gender
order by Customer_count desc;
```

Find out in which geography most customers are inactive.

```python
%sql
CREATE TABLE IF NOT EXISTS bank_analytics_project.gold.most_inactive_geography
USING DELTA
AS
SELECT Geography, count(distinct CustomerId) as Inactive_customer_count
FROM bank_analytics_project.silver.bank
WHERE Exited = 1
GROUP BY Geography
ORDER BY Inactive_customer_count DESC;
```

To attract young customers and make them not to exit, bank
wants to launch a scheme in a geography, so wants to find
geography wise how many customers are there with age less
than or equal to 35 and are active members.

```python
%sql
CREATE TABLE IF NOT EXISTS bank_analytics_project.gold.active_young_customers_by_geography
USING DELTA
AS
SELECT Geography, count(distinct CustomerId) as Active_young_customer_count
FROM bank_analytics_project.silver.bank
WHERE Age <= 35 AND Exited = 0
GROUP BY Geography
ORDER BY Active_young_customer_count DESC;
```

Find out the customers who are having 0 balance but still active
with the bank.

```python
%sql
CREATE OR REPLACE TABLE bank_analytics_project.gold.active_zero_balance_customers
USING DELTA AS
SELECT *
FROM bank_analytics_project.silver.bank
WHERE BalanceINR = 0 AND IsActiveMember = 'Yes';
```

Find out the customers whose est. salary is more than 12L
annually.

```python
%sql
CREATE OR REPLACE TABLE bank_analytics_project.gold.high_salary_customers AS
SELECT *
FROM bank_analytics_project.silver.bank
WHERE EstimatedSalaryINR > 1200000
```
